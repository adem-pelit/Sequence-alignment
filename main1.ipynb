```python
import numpy as np

# Hizalama işlemini yapacak olan fonksiyon...
def align(x,y,match=1, mismatch=-1, gap=-1):
    matris = {
        "x": x,
        "y": y,
        "liste": np.zeros((len(y)+1, len(x)+1),dtype=np.int32)
    }
    
    for i in range(len(matris["liste"])):
        matris["liste"][i][0] = i * gap

    for i in range(len(matris["liste"][0])):
        matris["liste"][0][i] = i * gap

    count = 0
    for i in range(len(y)):
        for j in range(len(x)):
            sutun, satir = i+1,j+1
            maxV = max([matris["liste"][sutun-1, satir-1], matris["liste"][sutun, satir-1], matris["liste"][sutun-1, satir]])

            if x[j] == y[i]:
                if matris["liste"][sutun-1, satir-1] == maxV:
                    matris["liste"][sutun, satir] = maxV + match
                else:
                    matris["liste"][sutun, satir] = maxV + gap
            else:
                if matris["liste"][sutun-1, satir-1] == maxV:
                    matris["liste"][sutun, satir] = maxV + mismatch
                else:
                    matris["liste"][sutun, satir] = maxV + gap
    path(matris)
    printS(matris)
    return matris

# matris uzerinden en uygun yolu hesaplayan fonksiyon...
def path(matris):
    pathlist = []
    sutun, satir = len(matris["y"]), len(matris["x"])
    loop = True
    while(loop):
        pathlist.append([sutun, satir])
        if sutun-1 < 0:
            if satir-1 < 0:
                loop = False
            else:
                sutun, satir = sutun, satir-1 
        else:
            if satir-1 < 0:
                sutun, satir = sutun-1, satir
            else:
                maxV = max([matris["liste"][sutun-1, satir-1], matris["liste"][sutun, satir-1], matris["liste"][sutun-1, satir]])
                if matris["liste"][sutun-1, satir-1] == maxV:
                    sutun, satir = sutun-1, satir-1
                elif matris["liste"][sutun, satir-1] == maxV:
                    sutun, satir = sutun, satir-1
                elif matris["liste"][sutun-1, satir] == maxV:
                    sutun, satir = sutun-1, satir
    matris["pathlist"] = pathlist
    matris["pathlist"].append([0,0])
    return pathlist

# Hizalama sonucunu cikti olarak veren fonksiyon    
def printS(matris):
    gapsChar = "_"
    pathlist,x,y = matris["pathlist"],matris["x"],matris["y"]
    pathlist.reverse()
    pathlist.pop(0)
    lastX = -1
    ust,orta,alt = [],[],[]
    for i in pathlist:
        if lastX != i[1]-1:
            ust.append(x[i[1]-1])
        else:
            ust.append(gapsChar)
        lastX = i[1]-1

    lastY = -1
    for i in pathlist:
        if lastY != i[0]-1:
            alt.append(y[i[0]-1])
        else:
            alt.append(gapsChar)
        lastY = i[0]-1
    for i in range(len(ust)):
        if ust[i] == alt[i]:
            orta.append("*")
        elif ust[i] == gapsChar or alt[i] == gapsChar:
            orta.append(" ")
        else:
            orta.append("|")
    matris["result"] = ''.join(str(e) for e in ust) + "\n" + ''.join(str(e) for e in orta) + "\n" + ''.join(str(e) for e in alt)

# hizalama matrisini yazdıran fonksiyon...
def printM(matris):
    print("\x1b[39m.\t.", end="\t")
    for i in matris["x"]:
        print(i, end="\t")
    print()
    for i in range(len(matris["liste"])):
        if i < 1:
            print(".",end="\t")
        else:
            print(matris["y"][i-1], end="\t")
        for j in range(len(matris["liste"][i])):
            if [i,j] in matris["pathlist"]:
                print("\x1b[31m{:3d}\x1b[39m".format(matris["liste"][i][j]),end="\t")
            else:
                print("{:3d}".format(matris["liste"][i][j]),end="\t")
        print()
    print()
```

```python
dna = align('GATTACA','GCATGCU',match=1, mismatch=-1, gap=-1)
printM(dna)
print(dna["result"],"\n")
```

```python
protein = align('KVLEFGY','EQLLKALEFKL',match=4, mismatch=-2, gap=-1)
printM(protein)
print(protein["result"],"\n")
```
